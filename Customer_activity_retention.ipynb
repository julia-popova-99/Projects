# Проект: Удержание активности постоянных клиентов интернет-магазина «В один клик». Обучение с учителем: качество модели

**Описание проекта:** Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.


**Цель исследования:** Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

**Задачи исследования:**

- Разработать модель, которая предскажет вероятность снижения покупательской активности.

- Выделить сегмент покупателей, проанализировать его и предложить, как увеличить его покупательскую активность. 

Возможные сегменты:

- Группа клиентов с максимальной долей покупок по акции и высокой вероятностью снижения покупательской активности.

- Группа клиентов, которые покупают только технику, то есть товары с длинным жизненным циклом.

- Группа клиентов, которые покупают товары из категории «Товары для себя» (новая категория, которую можно выделить на основе текущих) или «Товары для детей».

- Группа клиентов с высокой вероятностью снижения покупательской активности и наиболее высокой прибыльностью.

**Ход исследования:**

- Загрузка и изучение данных

- Предобработка данных

- Исследовательский анализ данных

- Объединение таблиц

- Корреляционный анализ

- Использование пайплайнов

- Анализ важности признаков

- Сегментация покупателей

**Общий вывод:** резюмирование полученных результатов, формулировка ключевых выводов.

**Описание данных**


- 1) market_file.csv

Таблица, о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.

<u>id</u> — номер покупателя в корпоративной базе данных.

<u>Покупательская активность</u> — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».

<u>Тип сервиса</u> — уровень сервиса, например «премиум» и «стандарт».

<u>Разрешить сообщать</u> — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.

<u>Маркет_актив_6_мес</u> — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.

<u>Маркет_актив_тек_мес</u> — количество маркетинговых коммуникаций в текущем месяце.

<u>Длительность</u> — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.

<u>Акционные_покупки</u> — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.

<u>Популярная_категория</u> — самая популярная категория товаров у покупателя за последние 6 месяцев.

<u>Средний_просмотр_категорий_за_визит</u> — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.

<u>Неоплаченные_продукты_штук_квартал</u> — общее число неоплаченных товаров в корзине за последние 3 месяца.

<u>Ошибка_сервиса</u> — число сбоев, которые коснулись покупателя во время посещения сайта.

<u>Страниц_за_визит</u> — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

- 2) market_money.csv

Таблица с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.

<u>id</u> — номер покупателя в корпоративной базе данных.

<u>Период</u> — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'. 

<u>Выручка</u> — сумма выручки за период. 

- 3) market_time.csv

Таблица с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода.

<u>id</u> — номер покупателя в корпоративной базе данных.

<u>Период</u> — название периода, во время которого зафиксировано общее время.

<u>минут</u> — значение времени, проведённого на сайте, в минутах.
- 4) money.csv

Таблица с данными о среднемесячной прибыли продавца за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.

<u>id</u> — номер покупателя в корпоративной базе данных.

<u>Прибыль</u> — значение прибыли.

## Использование пайплайнов

 - Подготовка данных

```python
X = df.drop(['покупательская_активность'], axis=1)
y = df['покупательская_активность']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=RANDOM_STATE, stratify = y)

display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

le = LabelEncoder()
le.fit_transform(y_train.unique())
print ('Значениям', le.classes_, 'назначено соответственно', le.transform(le.classes_))
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

Кодирование целевого признака:

'Прежний уровень' - 0

'Снизилась' - 1

Сформируем списки с названиями признаков

```python
ohe_columns = ['популярная_категория']
ord_columns = ['тип_сервиса', 'разрешить_сообщать']
```

```python
df['разрешить_сообщать'].unique()
```

```python
num_columns = ['выручка_препредыдущего_месяца',       
               'выручка_предыдущего_месяца',          
               'выручка_текущего_месяца',             
               'минут_предыдущего_месяца',            
               'минут_текущего_месяца',               
               'маркет_актив_тек_мес',
               'маркет_актив_6_мес',
               'длительность',                        
               'акционные_покупки',                   
               'средний_просмотр_категорий_за_визит',
               'неоплаченные_продукты_штук_квартал',  
               'ошибка_сервиса',                      
               'страниц_за_визит']
```

```python
# Создание пайплайна для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
    )
```

```python
# Создание пайплайна для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['стандарт', 'премиум'],
                    ['да', 'нет'],
                    
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough'
)
```

### Обучение модели: KNeighborsClassifier(), DecisionTreeClassifier(), LogisticRegression() и SVC().

   Для сравнения результатов работы обученных моделей необходимо выбрать универсальную метрику. В качестве такой метрики выступает ROC-AUC — показатель, который отражает истинную и стабильную способность модели к прогнозированию. Хорошая модель классификации имеет значение ROC-AUC > 0.9, и в идеале этот показатель должен стремиться к 1. При этом значение следует определять на тестовой выборке.

   + Кривая ROC демонстрирует частоту истинных положительных и ложных срабатываний при различных порогах классификации, в то время как AUC представляет собой обобщенную меру производительности модели машинного обучения по всем возможным порогам.
   + ROC расшифровывается как «Receiver Operating Characteristic» (кривая рабочих характеристик приемника). Это график, который иллюстрирует эффективность модели машинного обучения в задаче классификации, показывая частоту истинных срабатываний и частоту ложных срабатываний. AUC, в свою очередь, означает «Area Under the Curve» (площадь под кривой) и используется для измерения общей площади под кривой ROC.



```python
# Итоговый пайплайн
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', KNeighborsClassifier())
])

param_grid = [
    # Словарь для модели SVC()
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(1, 500),
        'models__max_features': range(1, 20),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [LogisticRegression(random_state=RANDOM_STATE, solver='liblinear',penalty='l1')],
        'models__C': range(1, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [SVC(random_state=RANDOM_STATE, probability=True)],
        'models__kernel': ['poly', 'rbf', 'sigmoid'],
        'models__degree': range(1, 15),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    }
]

model = RandomizedSearchCV(
                                pipe_final,
                                param_grid, 
                                cv=5,
                                scoring='roc_auc',
                                random_state=RANDOM_STATE,
                                n_jobs=-1
).fit(X_train, y_train)

print('Лучшая модель и её параметры:\n\n', model.best_estimator_)
print ('Метрика ROC-AUC лучшей модели на кросс - валидации:', model.best_score_)
```

```python
print('Площадь ROC-кривой на кросс-валидации:', model.best_score_)
roc_auc_test = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
print('Площадь ROC-кривой на тестовой выборке:', roc_auc_test)
roc_auc_km = (roc_auc_test - model.best_score_) / model.best_score_*100
print(f'Потеря качества на тестовой выборке  = {-roc_auc_km:.2f} %')
```

- Увеличение количества просмотренных страниц способствует увеличению времени, проведенного на сайте за визит. Если визит длится более 10 минут, вероятность сохранения покупательской активности значительно возрастает.

2) Клиенты с высокой прибыльностью и более 4 неоплаченных покупок в корзине

В данном сегменте также можно отметить значительное снижение покупательской активности. 

**Предложения:**
  
- Создать отдельную рабочую группу для этих клиентов с целью повышения их активности.
- Настроить систему уведомлений, сообщающую о наличии 6 и более неоплаченных товаров в корзине.  Такой подход может способствовать возврату пользователей на сайт, положительно повлиять на ключевые метрики и увеличить объемы продаж.

Ключевые рекомендации для обеих групп: сосредоточиться на продвижении акционных предложений, реализации рекламных кампаний, повышении вовлеченности на сайте и своевременном информировании о товарах в корзине, ожидающих оплаты.
