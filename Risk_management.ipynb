# Описание проекта: управление рисками и принятие решения о покупке коров для молочного хозяйства «Вольный луг»

**Заголовок:** Управление рисками и принятие решения о покупке коров для молочного хозяйства «Вольный луг».

**Описание проекта:** Фермер, владелец молочного хозяйства «Вольный луг» хочет купить бурёнок, чтобы расширить поголовье стада коров. Для этого он заключил выгодный контракт с ассоциацией пастбищ «ЭкоФерма».
Условия позволяют фермеру очень тщательно отобрать коров. Он определяет качество молока по строгой методике, и при этом ему нужно выполнять свой план развития молочного хозяйства. Фермер хочет, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, а её надой был вкусным — строго по его критериям, ничуть не хуже.

**Цель исследования:** Разработать модель машинного обучения, которая поможет управлять рисками и принимать объективное решение о покупке коров для молочного хозяйства «Вольный луг».

**Задачи исследования:**

- Создать 2 прогнозные модели для отбора коров в поголовье:

Модель 1 будет прогнозировать возможный удой коровы;
Модель 2 будет рассчитывать вероятность получить вкусное молоко от коровы.

- С помощью модели нужно отобрать коров по следующим критериям:

Средний удой за год — не менее 6000 килограммов;

Молоко должно быть вкусным.

**Ход исследования:**

- Загрузка и изучение данных

- Предобработка данных

- Исследовательский анализ данных

- Проведение корреляционного анализа признаков в датасете 

- Задача регрессии. Обучение трех модели простой линейной регрессии.

- Задача классификации

- Подготовка итоговых выводов

**Общий вывод:** резюмирование полученных результатов, формулировка ключевых выводов.

- 1) Файл ferma_main.csv содержит данные о стаде фермера на текущий момент. Описание данных:

<u>id</u> — уникальный идентификатор коровы.

<u>Удой, кг</u> — масса молока, которую корова даёт в год (в килограммах).

<u>ЭКЕ</u> (Энергетическая кормовая единица) — показатель питательности корма коровы.

<u>Сырой протеин, г</u> — содержание сырого протеина в корме (в граммах).

<u>СПО</u> (Сахаро-протеиновое соотношение) — отношение сахара к протеину в корме коровы.

<u>Порода</u> — порода коровы.

<u>Тип пастбища</u> — ландшафт лугов, на которых паслась корова.

<u>Порода папы_быка</u> — порода папы коровы.

<u>Жирность,%</u> — содержание жиров в молоке (в процентах).

<u>Белок,%</u> — содержание белков в молоке (в процентах).

<u>Вкус молока</u> — оценка вкуса по личным критериям фермера, бинарный признак (вкусно, не вкусно).

<u>Возраст</u> — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).


- 2) Файл ferma_dad.csv хранит имя папы каждой коровы в стаде фермера. Описание данных:

<u>id</u> — уникальный идентификатор коровы.

<u>Имя Папы</u> — имя папы коровы.


- 3) Файл cow_buy.csv — это данные о коровах «ЭкоФермы», которые фермер хочет изучить перед покупкой. Описание данных:

<u>Порода</u> — порода коровы.

<u>Тип пастбища</u> — ландшафт лугов, на которых паслась корова.

<u>Порода папы_быка</u> — порода папы коровы.

<u>Имя_папы</u> — имя папы коровы.

<u>Текущая_жирность,%</u> — содержание жиров в молоке (в процентах).

<u>Текущий_уровень_белок,%</u> — содержание белков в молоке (в процентах).

<u>Возраст</u> — возраст коровы, бинарный признак (менее_2_лет, более_2_лет).

## Обучение модели линейной регрессии

#### Модель линейной регрессии №1

**Отбор признаков**

Избавляемся от признака id, так как данный признак не является нужным для обучения модели. Также уберем признак "белок" в связи с почти отсутствием корреляции с целевым признаком. Целесообразно также убрать признак "жирность", являющийся характеристикой именно молока 

```python
ml = ferma_main.drop(['id', 'белок', 'жирность'], axis=1)
ml.columns
```

**Разбивка данных на выборки**

```python
RANDOM_STATE = 42

X = ml.drop('удой', axis=1)
y = ml['удой']

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE,
)
```

**Кодировка категориальных признаков с помощью OneHotEncoder.
Масштабирование количественных признаков с помощью StandardScaler.**

```python
# Разделение признаков на категориальные и количественные
cat_col = ['порода', 'тип_пастбища', 'порода_папы', 'вкус_молока', 'возраст']
num_col = ['эке', 'спо', 'сырой_протеин']

# Кодировка категориальных признаков
encoder = OneHotEncoder(drop='first', sparse_output=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col])
X_valid_ohe = encoder.transform(X_valid[cat_col])
encoder_col = encoder.get_feature_names_out()

# Масштабирование количественных признаков
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col])
X_valid_scaled = scaler.transform(X_valid[num_col])

# Собор выборок 
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col)
X_valid_ohe = pd.DataFrame(X_valid_ohe, columns=encoder_col)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col)
X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=num_col)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_valid = pd.concat([X_valid_ohe, X_valid_scaled], axis=1)
```

**Обучение модели**

```python
%%time
linreg_1 = LinearRegression()
linreg_1.fit(X_train, y_train)
predictions = linreg_1.predict(X_valid)
r2_1 = r2_score(y_valid, predictions)
mse_1 = mean_squared_error(y_valid, predictions)
rmse_1 = mean_squared_error(y_valid, predictions, squared=False)
mae_1 = mean_absolute_error(y_valid, predictions)
print(f'R2_score модели №1: {r2_1}')
```

**Анализ остатков**

```python
# Выделение остатков
residuals_1 = y_valid - predictions

# Построение гистограммы распределения
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].hist(residuals_1, bins=50, color='blue', edgecolor='gray')
zero = axes[0].vlines(x=0, colors='dimgrey', ymin=0, ymax=11, linestyles='--', label='Средняя сумма остатков Модели №1')
median = axes[0].vlines(x=residuals_1.median(), colors='red', ymin=0, ymax=11, label='Смещение медианы')
axes[0].legend()
axes[0].set_title('Гистограмма распределения остатков Модели №1')
axes[0].set_xlabel('Остатки показатели')

# Построение графика дисперсии
axes[1].scatter(predictions, residuals_1, color='purple', alpha=0.4)
axes[1].set_xlabel('Предсказания модели')
axes[1].set_ylabel('Остатки показатели')
axes[1].set_title('Дисперсия Модели №1')
plt.show()
```

Выводы по Модели №1:

r2_score Модели №1 == 0.7951757980652822

Смешение медианы предсказаний незначительное. Остатки имеют нормальное распределение. 

Остатки имеют постоянную дисперсию на интервале использования Модели №1. Самый большой разброс в остатках имеет интервал с значениях от 5500 от 6500, что может указывать на неточное предсказание целевого признака. С целью улучшения модели необходимо конкретизировать данные по данному интервалу.

```python
# Функция для расчета метрик для каждого порога
def choice_threshold(threshold_list, proba_df):
    result_df = pd.DataFrame(columns=['порог', 'precision', 'recall', 'колич_FP', 'общее_предсказанных_объектов',
                                      'объекты_класса_1'])

    for threshold in threshold_list:
        column_name = 'y_pred_' + str(threshold)

        # Precision
        precision = proba_df.loc[proba_df[column_name] == 1, 'y_valid'].mean()

        # Число предсказанных объектов
        predicted_objects = proba_df[column_name].sum()

        # Всего объектов класса 1
        total_objects_class_1 = proba_df['y_valid'].sum()

        # Recall
        recall = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 1)].shape[
                     0] / total_objects_class_1

        # Количество ложноположительных предсказаний (FP)
        false_positives = proba_df.loc[(proba_df[column_name] == 1) & (proba_df['y_valid'] == 0)].shape[0]

        # Добавляем результаты в датафрейм
        result_df = result_df.append({
            'порог': threshold,
            'precision': precision,
            'recall': recall,
            'колич_FP': false_positives,
            'общее_предсказанных_объектов': predicted_objects,
            'объекты_класса_1': total_objects_class_1,
        }, ignore_index=True)

    return result_df
```

```python
# Перебор порогов
proba_data_1 = data_proba.copy(deep=True)

thresholds_1 = [round(i, 2) for i in np.linspace(0.1, 1, num=10, endpoint=False)]
columns = []

for i in thresholds_1:
    columns.append('y_pred_' + str(i))
    proba_data_1['y_pred_' + str(i)] = proba_data_1['y_proba'].apply(lambda x: 1 if x > i else 0)

proba_data_1.head()
```
