# Описание проекта: исследование игрового рынка. 

**Заголовок:** Исследование игрового рынка.

**Описание проекта:** Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Представленные данные до 2016 года. Планируемая кампания - 2017 год. 

Аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Цель исследования:** Определение закономерностей успешности игры для выделения потенциально популярного продукта с целью планирования рекламных кампаний. 

**Ход исследования:**

- Изучение общей информации

- Подготовка данных

- Проведение исследовательского анализа данных

- Составление портрета пользователя каждого региона

- Проверка гипотез

**Общий вывод:** резюмирование полученных результатов, формулировка ключевых выводов.

**Описание данных**

<u>Name</u> — название игры

<u>Platform</u> — платформа

<u>Year_of_Release</u> — год выпуска

<u>Genre</u> — жанр игры

<u>NA_sales</u> — продажи в Северной Америке (миллионы проданных копий)

<u>EU_sales</u> — продажи в Европе (миллионы проданных копий)

<u>JP_sales</u> — продажи в Японии (миллионы проданных копий)

<u>Other_sales</u> — продажи в других странах (миллионы проданных копий)

<u>Critic_Score</u> — оценка критиков (максимум 100)

<u>User_Score</u> — оценка пользователей (максимум 10)

<u>Rating</u> — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Данные за 2016 год могут быть неполными.

# Подготовка данных

#### Замена названий столбцов

#### Преобразование данных в нужные типы и обработка пропусков

```python
# Приведение в столбце user_score значений 'tbd' на NaN, так как
# 'tbd' - to be determined - используется, когда что-то еще не решено или пока неизвестно
# Изменение типа данных в user_score на float
games['user_score'] = games['user_score'].replace('tbd', np.NaN)
games['user_score'] = games['user_score'].astype('float')
```

```python
# Подсчет пропусков в столбцах
games.isna().sum().sort_values(ascending=False)
```

Больше всего пропусков в столбцах user_score, critic_score, rating. 

В столбце rating (рейтинг от ESRB) заполним пропуски на 'unknown', так как ни медианным, ни средним значением пропуски не заполнит. Появление данных пропусков может объясняться датой выходя игры. Вероятно ассоциация ESRB еще не существовала в тот период.

Также в столбце user_score присутствовало значение 'tbd' (уже устранили в предыдущем шаге), означающее, что оценка пользователей еще не решена или пока неизвестна (то есть возможно заменить 'tbd' на NaN, так как по факту это неизвестное значение). В столбце user_score больше всего пустых значений, тип 'object'. Зполним пропуски маркерным значением (-1), чтобы не повлиять на общую статистику. Пропуски возможны по причине того, что игроками еще не была проставлена оценка, не было подходящего сервиса для проставления оценок.

В столбце critic_score тоже заменим пропуски маркерным значением (-1). Пропуски возможны по причине того, что критиками еще не была проставлена оценка, не было подходящего сервиса для проставления оценок.

В столбце year_of_release 269 пропусков, также заменим их маркерным значением (-1). Природа данных пропусков неизвестна.

# Проведение исследовательского анализ данных

#### Количество игр, выпущенных в разные годы

```python
# Функция для создания столбчатых диаграмм по данным
def graph_bar(data, yaxes, xlab, ylab, title):
    plt.figure(figsize=(10,6))
    ax = sns.barplot(x = data.index, y = yaxes, data = data)
    ax.set(xlabel = xlab, ylabel = ylab, title = title)
    plt.xticks(rotation = 45)
    plt.show()
```

```python
games_years_release = games[games['year_of_release'] > 0].pivot_table(index = 'year_of_release',
        values = 'name', aggfunc = 'count').sort_values('year_of_release', ascending = False)
graph_bar(games_years_release, 'name', 'Год выпуска', 'Количество игр', 'Количество выпущенных игр по годам')
```

До 1994 года выпускалось небольшое количество игр в год.


На графике изображен резкий скачок выпуска видеоигр с 1995 года. Данный скачок вызван выходом приставки PlayStation 1. 

Второй скачок заметен с 2002 года. Причиной послужил выход PlayStation 2, выход приставок Microsoft c Xbox, рост популярности PC-платформ.

Пиковые значения наблюдаются в 2008-2009 годах. Далее значения идут на спад, вероятно по причине развития мобильников и социальных сетей. Это могло снизить интерес к видеоиграм.

#### Количество выпущенных игр по платформам

```python
platforms_count = games[games['year_of_release'] >= 1994].pivot_table(index = 'platform',
      values = 'name', aggfunc = 'count').sort_values('name', ascending = False)
graph_bar(platforms_count, 'name', 'Количество игр', 'Название платформы', 'Количество выпущенных игр по платформам')
```

Лидером являются PlayStation 2 и Nintendo DS. Далее идут с значительным отставанием PlayStation 3, Wii, Xbox 360,. 

Самые редкие - платформы японского происхождения.

#### Лидирующие платформы по продажам. 

```python
platforms_sales=games[games['year_of_release'] >= 1994].pivot_table(index = 'platform',
      values ='sales_total', aggfunc = 'sum').sort_values('sales_total', ascending=False)
graph_bar(platforms_sales, 'sales_total', 'Название платформы', 'Количество продаж (млн. копий)', 'Количество игр на платформе за все время')
```

На графике заметны явные 6 лидеров: PS2, X360, PS3, Wii, DS, PS.

Рассмотрим данные платформы на промежутках времени. 

```python
# Функция для создания столбчатых диаграмм
def graph_line(data, xaxes, yaxes, xlab, ylab, title):
    plt.figure(figsize=(10,6))
    ax = sns.lineplot(x = xaxes, y = yaxes, data = data, hue = 'platform')
    ax.set(xlabel = xlab, ylabel = ylab, title = title)
    plt.show()
```

```python
# Создание списка лидеров по продажам
platforms_top = platforms_sales.head(6).index.tolist()
# Создание сводной таблицы лидирующих платформ по продажам в зависимости от года
platforms_top_sales = games[(games['platform'].isin(platforms_top)) & (games['year_of_release'] >= 1994)].pivot_table(
    index = ['platform','year_of_release'], values = 'sales_total', aggfunc = 'sum').reset_index()
# Создание графика продаж в зависимости от года по каждой лидирующей платформе
graph_line(platforms_top_sales, 'year_of_release', 'sales_total', 'Год',
           'Количество продаж (млн. копий)', 'Топ 6 платформ по продажам в зависимости от года')
```

На графике заметно, что в среднем платформа активно функционирует около 7-10 лет, далее идет существенный спад количества продаж.

Платформа PS начала рост с 1995 года по 2003 год.

Платформа PS2 c 2000 по 2010 год.

Данные по платформе DS начали собирать после начала ее функционирования, поэтому данные отображаются на графике с 2004 года. С 2004 года идет рост, пик 2007 год, спад до 2013 года.

Данные по платформе Wii так же начали собирать после начала ее функционирования, поэтому данные отображаются на графике с 2005 года. С 2005 года идет рост, пик 2008 год, спад до 2015 года.

Платформа PS3 рост наблюдается с 2006 года, пик 2011 год, спад влоть до 2014 года.

Платформа X360 активно начала рост с 2005 года, спад продолжался до 2015 года. Пик продаж пришелся на 2011 год.

Согласно данным лидером продаж за последние 3 года является PS4. Создадим диаграмму рассеяния, отображающую отзывы покупателей (зеленые точки) и критиков (синие точки). 

```python
# Функция для создания графиков корреляции
def graph_scatter(data, platform):
    plt.figure(figsize=(10,6))
    ax = sns.scatterplot(x='sales_total', y='critic_score', data = data, alpha = 0.5, color='blue',)
    ax = sns.scatterplot(x='sales_total', y='user_score', data = data, alpha = 0.2, color='green');
    ax.set(xlabel = 'Продажи (млн. копий)', ylabel = 'Оценки', title = platform)
    plt.show()
    print(f'Корреляция продаж и отзывов критиков "{platform}":',
        round(data['sales_total'].corr(data['critic_score']),3))
    print(f'Корреляция продаж и отзывов пользователей "{platform}":',
        round(data['sales_total'].corr(data['user_score']),3))
```

# Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

<u>Нулевая гипотеза (Hо):</u> средние пользовательские рейтинги платформ Xbox One и PC одинаковые

<u>Альтернативная гипотеза (H1):</u> средние пользовательские рейтинги платформ Xbox One и PC отличаются

```python
pc_user_score = games[(games['platform'] == 'PC') & (games['user_score'] >= 0) & (games['year_of_release'] >= 2014)]['user_score']
xone_user_score = games[(games['platform'] == 'XOne') & (games['user_score'] >= 0)]['user_score']
# Проверка количества значений в выборках
print('Количество значений в выборке для платформы Xbox One:', xone_user_score.shape[0])
print('Количество значений в выборке для платформы PC:', pc_user_score.shape[0])
```

Для проверки гипотезы применим t-тест. Перед нами 2 независимые выборки. 2 выборки разного размера. Зададим параметр equal_var = False, так как скорее всего дисперсии не равны.

```python
alpha = 0.05

result = st.ttest_ind(pc_user_score, xone_user_score, equal_var = False)

print('Значение P-value ', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем гипотезу Hо')
else:
    print('Не можем отвергнуть гипотезу Hо')
```

```python
# Проверим средние значения по выборкам
print('Среднее значение пользовательского рейтинга платформы PC:', round(pc_user_score.mean(),2))
print('Среднее значение пользовательского рейтинга платформы Xbox One:', round(xone_user_score.mean(),2))
```

В ходе проверки гипотезы мы пришли к выводу, что средние пользовательские рейтинги платформ (за актуальный период) Xbox One и PC одинаковые.


#### Средние пользовательские рейтинги жанров Action и Sports разные.

<u>Нулевая гипотеза (H0):</u> средние пользовательские рейтинги жанров Action и Sports равны

<u>Альтернативная гипотеза (H1):</u> средние пользовательские рейтинги жанров Action и Sports отличаются

```python
аction_user_score = games[(games['genre'] == 'Action') & (games['year_of_release'] >= 2014)]['user_score']
sports_user_score = games[(games['genre'] == 'Sports') & (games['year_of_release'] >= 2014)]['user_score']
# Проверка количества значений в выборках
print('Количество значений в выборке для жанра Action:', аction_user_score.shape[0])
print('Количество значений в выборке для жанра Sports:', sports_user_score.shape[0])
```

Для проверки гипотезы применим t-тест. Перед нами 2 независимые выборки. 2 выборки разного размера. Зададим параметр equal_var = False, так как скорее всего дисперсии не равны.

```python
alpha = 0.05

result = st.ttest_ind(аction_user_score, sports_user_score, equal_var = False)

print('Значение P-value ', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем гипотезу H0')
else:
    print('Не можем отвергнуть гипотезу H0')
```

```python
# Проверим средние значения по выборкам
print('Среднее значение пользовательского рейтинга жанра Action:', round(аction_user_score.mean(),2))
print('Среднее значение пользовательского рейтинга жанра Sports:', round(sports_user_score.mean(),2))
```

В ходе проверки гипотезы мы пришли к выводу, что средние пользовательские рейтинги жанров Action и Sports отличаются.
